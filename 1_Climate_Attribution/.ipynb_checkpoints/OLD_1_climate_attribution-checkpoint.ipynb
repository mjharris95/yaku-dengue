# Load GCM data

```python
pr = {}
for kk, model in enumerate(model_list):
    print('loading data from', model, '...')
    if kk not in model_skip_list:
        pr[model] = {}
        for ll, var in enumerate(variant_list[model]):
            time_bnds = slice('1850-01-01', '2015-01-01')
            pr_files = sorted(glob.glob('input_data/CMIP6/pr_Amon_'+model+'_historical_'+var+'*.nc'))

            ## check for entire 1850-2014 file in list ##
            if (len(pr_files) > 0):
                pr_f_idx = np.argwhere(['185001-201412' in pr_files[i] for i in range(len(pr_files))])
                if sum(pr_f_idx) > 0:
                    pr_files = np.array(pr_files)[pr_f_idx[0]]
            
            pr[model][var] = xr.open_mfdataset(pr_files).pr.sel(lat=region_lat, lon=region_lon)*86400 # kg m-2 s-1 --> mm / day
            weights = np.cos(np.deg2rad(pr[model][var].lat))
            weights.name = "weights"
            pr[model][var] = pr[model][var].weighted(weights).mean(dim=['lat', 'lon']).sel(time=time_bnds)
            march_idx = pr[model][var].time.dt.month == 3
            pr[model][var] = pr[model][var].isel(time=march_idx)

pr_vals = {}
for kk, model in enumerate(model_list):
    print('accessing', model, 'data ...')
    if kk not in model_skip_list:         
        pr_vals[model] = {}
        for ll, var in enumerate(variant_list[model]):
            pr_vals[model][var] = pr[model][var].values
```

# Has anthropogenic forcing increased the frequency of extreme March precipitation (above 85th percentile)?

```python
## calculate 85th percentile threshold from the period 1985-2014 ##
qPR = 85
pr_85 = {}
for jj, model in enumerate(model_list):
    if jj not in model_skip_list: 
        pr_85[model] = {}
        for var in variant_list[model]:
            last_30_yrs_idx = (pr[model][var].time.dt.year >= 1985) & (pr[model][var].time.dt.year <= 2014)
            pr_85[model][var] = np.percentile(pr_vals[model][var][last_30_yrs_idx], q=qPR)

model = model_list[0]
var = variant_list[model_list[0]][0]
wet_idx = np.ones([len(pr.keys())*len(variant_list[model]), len(pr_vals[model][var])]).astype(int)
ct = -1
for jj, model in enumerate(model_list):
    if jj not in model_skip_list: 
        for var in variant_list[model]:
            ct = ct+1
            wet_idx[ct,:] = (pr_vals[model][var] >= pr_85[model][var]).astype(int)
```

```python
fs = 16
model = model_list[0]
var = variant_list[model_list[0]][0]

fig, ax = plt.subplots(1,2,figsize=(15,8), gridspec_kw={'width_ratios': [4, 1], 'wspace':0.0})
fraction = 100*np.mean(wet_idx, axis=0)
early_idx = pr[model][var].time.dt.year < 1900
mid_idx = (pr[model][var].time.dt.year >= 1900) & (pr[model][var].time.dt.year < 1965)
late_idx = pr[model][var].time.dt.year >= 1965

title_list = ['P'+str(qPR)]
alph=0.6
lw=2.5

## Plot timeseries of  ##
ax[0].scatter(np.arange(1850,1900), fraction[early_idx], color='mediumblue', alpha=alph, s=50, lw=0)
ax[0].scatter(np.arange(1900,1965), fraction[mid_idx], color='black', alpha=alph, s=50, lw=0)
ax[0].scatter(np.arange(1965,2015), fraction[late_idx], color='tab:red', alpha=alph, s=50, lw=0)
ax[0].hlines(np.mean(fraction[early_idx]), xmin=1850, xmax=1899, color='mediumblue', ls='solid', lw=4, alpha=alph+0.15)
ax[0].hlines(np.mean(fraction[mid_idx]), xmin=1900, xmax=1965, color='black', ls='solid', lw=4, alpha=alph+0.15)
ax[0].hlines(np.mean(fraction[late_idx]), xmin=1965, xmax=2014, color='tab:red', ls='solid', lw=4, alpha=alph+0.15)
ymax=ax[0].get_ylim()[1]+1.1
ax[0].set_xlim(1849,2015)
ax[0].set_ylim(3,ymax)
ax[0].tick_params(axis='both', labelsize=fs+1)
ax[0].set_ylabel('percent of climate model simulations\nexceeding 85th percentile (%)', fontsize=fs+4)
ax[0].set_xlabel('year', fontsize=fs+4)


## Plot box faces ##
wid=0.14
lw=1.4
ax[1].boxplot(fraction[early_idx], positions=[0.008], widths=wid,
           labels=['1850-1899'],patch_artist=True, showfliers=False,
           medianprops={"color":'black', "alpha":1, "lw":lw},
           capprops={"color":'black', "alpha":1, "lw":lw},
           whiskerprops={"color":'black', "alpha":1, "lw":lw},
           boxprops={"facecolor":'mediumblue', "alpha":alph})
ax[1].boxplot(fraction[mid_idx], positions=[0.17], widths=wid,
           labels=['1900-1964'],patch_artist=True, showfliers=False,
           medianprops={"color":'black', "alpha":1, "lw":lw},
           capprops={"color":'black', "alpha":1, "lw":lw},
           whiskerprops={"color":'black', "alpha":1, "lw":lw},
           boxprops={"facecolor":'black', "alpha":alph})
ax[1].boxplot(fraction[late_idx], positions=[0.332], widths=wid,
           labels=['1965-2014'],patch_artist=True, showfliers=False,
           medianprops={"color":'black', "alpha":1, "lw":lw},
           capprops={"color":'black', "alpha":1, "lw":lw},
           whiskerprops={"color":'black', "alpha":1, "lw":lw},
           boxprops={"facecolor":'tab:red', "alpha":alph})

## Plot box borders ##
ax[1].boxplot(fraction[early_idx], positions=[0.008], widths=wid,
           labels=['1850-1899'],patch_artist=False, showfliers=False,
           medianprops={"color":'black', "alpha":0, "lw":lw},
           boxprops={"color":'black', "alpha":1, "lw":lw})
ax[1].boxplot(fraction[mid_idx], positions=[0.17], widths=wid,
           labels=['1900-1964'],patch_artist=False, showfliers=False,
           medianprops={"color":'black', "alpha":0, "lw":lw},
           boxprops={"color":'black', "alpha":1, "lw":lw})
ax[1].boxplot(fraction[late_idx], positions=[0.332], widths=wid,
           labels=['1965-2014'],patch_artist=False, showfliers=False,
           medianprops={"color":'black', "alpha":0, "lw":lw},
           boxprops={"color":'black', "alpha":1, "lw":lw})

ax[1].set_xlim(-0.1,0.5)
ax[1].set_ylim(3,ymax)
ax[1].tick_params(axis='both', labelsize=fs+1)

## K-S Test ##
pval = ks_2samp(fraction[early_idx], fraction[mid_idx]).pvalue
print('Early-Mid P-value =', pval)
xadj=-0.02
if pval < 0.001:
    ax[1].annotate('p$\,$<$\,$0.001', xy=(0.66-xadj,0.78), color='black', alpha=alph+0.15,
                xycoords='axes fraction', ha='right', va='top', fontsize=fs+1, fontweight='bold')
else:
    ax[1].annotate('p$\,$=$\,$'+'{:.3f}'.format(np.round(pval,5))+'', xy=(0.66-xadj,0.78), color='black',alpha=alph+0.15,
                xycoords='axes fraction', ha='right', va='top', fontsize=fs+1, fontweight='bold')
    
pval = ks_2samp(fraction[early_idx], fraction[late_idx]).pvalue
print('Early-Late P-value =', pval)
if pval < 0.001:
    ax[1].annotate('p$\,$<$\,$0.001', xy=(0.91-xadj,0.91), color='tab:red',alpha=alph+0.15,
                xycoords='axes fraction', ha='right', va='top', fontsize=fs+1, fontweight='bold')
else:
    ax[1].annotate('p$\,$=$\,$'+'{:.3f}'.format(np.round(pval,5))+'', xy=(0.91-xadj,0.91), color='red',alpha=alph+0.15,
                xycoords='axes fraction', ha='right', va='top', fontsize=fs+1, fontweight='bold')
    
pval = ks_2samp(fraction[mid_idx], fraction[late_idx]).pvalue
print('Mid-Late P-value =',pval)

model_explain_str = ''

if region_str == 'northwest_peru':
    ax[0].set_title('Frequency of extreme March monthly precipitation in northwestern Peru'+model_explain_str, 
                 fontsize=fs+6.0, fontweight='bold',x=0.56,y=1.03)
else:
    fig.suptitle('Frequency of extreme March monthly precipitation in Peru'+model_explain_str, 
                 fontsize=fs+6.5, fontweight='bold', y=0.95)
        
## Period Means ##
print('early mean =',np.mean(fraction[early_idx]))
print('mid mean =',np.mean(fraction[mid_idx]))
print('late mean =',np.mean(fraction[late_idx]))

adj=0.04
xadj = 0.01
xx = 0.17
ax[0].annotate('                                    \n\n\n\n', xy=(0.038+xadj,0.997-adj), color='black', alpha=1,
                bbox = dict(boxstyle='round', facecolor='gainsboro', alpha=0.4, lw=0.4),
                xycoords='axes fraction', ha='left', va='top', fontsize=fs+4.8)
ax[0].annotate('CMIP6 period mean:', xy=(0.039+xadj,0.985-adj), color='black', alpha=1,
                xycoords='axes fraction', ha='left', va='top', fontsize=fs+4.5, fontweight='semibold')
ax[0].annotate('1850-1899', xy=(0.05+xadj+xx,0.92-adj), color='mediumblue', alpha=alph+0.25,
                xycoords='axes fraction', ha='right', va='top', fontsize=fs+4.5)
ax[0].annotate('1900-1964', xy=(0.05+xadj+xx,0.86-adj), color='black', alpha=alph+0.25,
                xycoords='axes fraction', ha='right', va='top', fontsize=fs+4.5)
ax[0].annotate('1965-2014', xy=(0.05+xadj+xx,0.80-adj), color='tab:red', alpha=alph+0.25,
                xycoords='axes fraction', ha='right', va='top', fontsize=fs+4.5)
ax[0].annotate(' = '+str('{:.2f}'.format(np.mean(fraction[early_idx])))+'%', xy=(0.05+xadj+xx,0.92-adj), color='k', alpha=1,
                xycoords='axes fraction', ha='left', va='top', fontsize=fs+4.5)
ax[0].annotate(' = '+str('{:.2f}'.format(np.mean(fraction[mid_idx])))+'%', xy=(0.05+xadj+xx,0.86-adj), color='k', alpha=1,
                xycoords='axes fraction', ha='left', va='top', fontsize=fs+4.5)
ax[0].annotate(' = '+str('{:.2f}'.format(np.mean(fraction[late_idx])))+'%', xy=(0.05+xadj+xx,0.80-adj), color='k', alpha=1,
                xycoords='axes fraction', ha='left', va='top', fontsize=fs+4.5)
ax[1].axis('off')

right_ax = ax[0].secondary_yaxis("right")
right_ax.set_yticks(ax[0].get_yticks())
right_ax.set_yticklabels([])
ax[0].tick_params(axis='both', which='major', length=9, direction='inout')#, labelsize=12)    
right_ax.tick_params(axis='both', which='major', length=9, direction='inout')

plt.savefig('figures/extreme_March_precip.jpg', dpi=500, bbox_inches='tight', pad_inches=0.2)
plt.savefig('figures/extreme_March_precip.pdf', dpi=500, bbox_inches='tight', pad_inches=0.2)
print("Change in 1965-2014 mean relative to 1850-1899 =",100*(np.mean(fraction[late_idx]) - np.mean(fraction[early_idx]))/ np.mean(fraction[early_idx]), '%')
print("Change in 1900-1964 mean relative to 1850-1899 =",100*(np.mean(fraction[mid_idx]) - np.mean(fraction[early_idx]))/ np.mean(fraction[early_idx]), '%')
```
